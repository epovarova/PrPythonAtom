```python
r = requests.get("https://wikiway.com/dostoprimechatelnosti/top100/")
print(r.ok)
print(r.status_code)
print(r.cookies)
soup = BeautifulSoup(r.text, "html.parser")
```

```python
href =[]
for i in soup.findAll("div", attrs={"class": ["item-text"]}):
    href.append('https://wikiway.com' + i.find('a')['href'])
print (href[0])
```

```python
req = requests.get(href[1])
print(req.ok)
print(req.status_code)
print(req.cookies)
soup1 = BeautifulSoup(req.text, "html.parser")
```

```python
dictionary = {}
for ind in range(len(href)):
    req = requests.get(href[ind])
#print(req.ok)
#print(req.status_code)
#print(req.cookies)
    soup1 = BeautifulSoup(req.text, "html.parser")

    s=''
    title = soup1.findAll("div", attrs={"class": ["contents"]})[0].find('strong').contents[0]
    #print (title)
    for i in soup1.findAll("div", attrs={"class": ["contents"]}):
        for j in i.findAll('p'):        
            for q in j.contents:
                s = s + str(q)
            s = s + "\n"
        s = re.sub(r"<[^>]*>","",s)
        s = s.strip()
        s = s[0:s.rfind("\n")]
        dictionary[title] = s
#print (dictionary)
```

```python
import sys
sys.setrecursionlimit(100000)
```

```python
import pickle
with open('parser_data.txt','wb') as out:
    pickle.dump(dictionary,out)

with open('parser_data.txt','rb') as inp:
    d_in = pickle.load(inp)
print(d_in)
```
